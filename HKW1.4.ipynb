```python
import yfinance as yf
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
start_data='1990-01-01'
```

```python
end_data='2021-07-07'
```

```python
tickers_list=['NVDA','AMD','MSFT']
```

```python
data=pd.DataFrame(columns=tickers_list)
```

```python
for ticker in tickers_list:
    data[ticker]=yf.download(ticker,start_data,end_data)['Adj Close']
```

```python
data.head()
```

```python
data.plot(figsize=(10,7))
```

```python
plt.legend()
```

```python
plt.title("Adjust Close Price",fontsize=16)
```

```python
plt.ylabel('Price',fontsize=14)
```

```python
plt.xlabel('Year',fontsize=14)
```

```python
plt.grid(which="major",color='k',linestyle='-',linewidth=0.5)
```

```python
plt.show()
```
